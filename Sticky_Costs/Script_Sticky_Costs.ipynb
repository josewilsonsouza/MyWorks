```python
from pystoned import DEA
from pystoned import dataset as dataset
from pystoned.constant import RTS_VRS, OPT_LOCAL

# import the GHG example data
data = dataset.load_GHG_abatement_cost()

# define and solve the DEA DDF model
model = DEA.DDF(y=data.y, x=data.x, b=data.b, gy=[1], gx=[0.0, 0.0], gb=[-1], rts=RTS_VRS, yref=None, xref=None, bref=None)
model.optimize(OPT_LOCAL)

# display the technical efficiency
model.display_theta()

# display the intensity variables
model.display_lamda()
```

```python
def loc_emp(df, emp):
    return df.loc[df.EMPRESA == emp]
```

```python
def diff_log(df, var, log = 'log'):
    '''
    Função que calcula a diferença logaritima da variável inserida e aplica no painel.
    '''
    df_concat = pd.DataFrame()
    for emp in df.EMPRESA.unique():
        df_emp = loc_emp(df, emp).copy()
        
        dlog = []
        diff_log10 = np.log10(df_emp[var]) - np.log10(df_emp[var].shift(1))
        diff_ln = np.log(df_emp[var]) - np.log(df_emp[var].shift(1))
        
        for dlg, dln in zip(diff_log10, diff_ln):
            if log == 'log':
                if var == 'RLV':
                    if abs(dlg) > np.log10(2):
                        dlog.append(np.nan)
                    else:
                        dlog.append(dlg)
                else:
                    dlog.append(dlg)
            
            elif log == 'ln':
                if var == 'RLV':
                    if abs(dlg) > np.log10(2):
                        dlog.append(np.nan)
                    else:
                        dlog.append(dln)
                else:
                    dlog.append(dln)
            
        df_emp[f'diff_{log}_{var}'] = dlog
            
        df_concat = pd.concat([df_concat, df_emp])
    
    return df_concat
```

```python
def reg_fixed_effects(df, list_dep_vars, indep_var, units):
    '''
    Função para aplicação do modelo de regressão linear.
    Entrada.
        df: dataframe
        list_dep_vars: Variáveis Eplicativas
        indep_var: Variável Depedente
        units: Unidades
    Saída.
        Estatisticas do Modelo
    '''
    df_reg1 = df[list_dep_vars+[indep_var, units]].copy()
    df_reg1 = df_reg1.dropna()

    dependents_vars = '+'.join(list_dep_vars)

    y,X = dmatrices(f'{indep_var} ~ {dependents_vars}+C({units})', df_reg1, return_type='dataframe')

    model = sm.OLS(y, X).fit()
    print_model = model.summary()

    return print(print_model)
```

```python
reg_fixed_effects(df, ['ln_RLV', 'D_t_x_dLog_RLV'], 'dLog_CPV','EMPRESA')
```
